¿Cuáles son los colegios que ofrecen todos los niveles educativos? ¿Qué idiomas ofrecen?

```python
df_colegios.niveles.unique()
```

```python
# Definir el conjunto de todos los niveles educativos
niveles_educativos_completos = {'PREESCOLAR', 'BÁSICA PRIMARIA', 'BÁSICA SECUNDARIA', 'MEDIA', 'PRIMERA INFANCIA'}

# Función para determinar si un colegio ofrece todos los niveles educativos
def ofrece_todos_los_niveles(niveles):
    if pd.isna(niveles):  # Manejar casos donde el valor es NaN
        return False
    niveles_set = set(niveles.split(','))
    return niveles_educativos_completos.issubset(niveles_set)

# Filtrar el DataFrame para obtener solo los colegios que ofrecen todos los niveles
colegios_todos_niveles = df_colegios[df_colegios['niveles'].apply(ofrece_todos_los_niveles)]
colegios_todos_niveles
```

```python
idiomas_ofrecidos = colegios_todos_niveles['idiomas'].dropna().unique()

print(idiomas_ofrecidos)
```

d) ¿Cuál era la tasa de desocupación en el período seleccionado para las mujeres? ¿Era más alta o más baja que la de los hombres? (Extra: pensar el resultado usando una función)

Nota: la tasa se calcula como Desocupados / (Ocupados + Desocupados)

```python
eph_2['sexo'] = eph_2['sexo'].replace({'Mujer': 2, 'Hombre': 1})


def calcular_tasa_desocupacion(df, sexo):
    # Filtrar por sexo
    df_sexo = df[df['sexo'] == sexo]

    # Contar desocupados y ocupados
    desocupados = df_sexo[df_sexo['estado'] == 2].shape[0]
    ocupados = df_sexo[df_sexo['estado'] == 1].shape[0]

    # Calcular tasa de desocupación
    if ocupados + desocupados == 0:  # Evitar división por cero
        return None
    tasa_desocupacion = desocupados / (ocupados + desocupados)
    return tasa_desocupacion

# Calcular la tasa de desocupación para mujeres (2) y hombres (1)
tasa_desocupacion_mujeres = calcular_tasa_desocupacion(eph_2, 2)
tasa_desocupacion_hombres = calcular_tasa_desocupacion(eph_2, 1)

print(f"Tasa de desocupación para mujeres: {tasa_desocupacion_mujeres:.2%}")
print(f"Tasa de desocupación para hombres: {tasa_desocupacion_hombres:.2%}")

# Comparar las tasas
if tasa_desocupacion_mujeres > tasa_desocupacion_hombres:
    print("La tasa de desocupación para las mujeres es más alta que para los hombres.")
elif tasa_desocupacion_mujeres < tasa_desocupacion_hombres:
    print("La tasa de desocupación para las mujeres es más baja que para los hombres.")
else:
    print("La tasa de desocupación para mujeres y hombres es igual.")
```
